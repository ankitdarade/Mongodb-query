```python
#importing libraries
import csv
import json
from pymongo import MongoClient
```

```python
# Connecting to MongoDB
client = MongoClient('localhost', 27017)
db = client['Exponentdata']
collection = db['Whetherdata']
```

```python
# Open the CSV
with open(r"D:\python\datanew.csv", 'r') as f:
    reader = csv.DictReader(f)
    rows = list(reader)
```

```python
# Convert to JSON
json_data = json.dumps(rows)
```

```python
# Insert into MongoDB
collection.insert_many(json.loads(json_data))
```
